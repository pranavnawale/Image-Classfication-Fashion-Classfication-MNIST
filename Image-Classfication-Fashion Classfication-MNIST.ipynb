# Image-Classfication-Fashion Classfication-MNIST

```python
# Importing the Dependencie
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
import seaborn as sns 
from matplotlib import pyplot as plt
from keras import Sequential
```

```python
# checking the version of tensorflow
print(tf.__version__)
```

```python
# Load Data
(X_train,y_train),(X_test,y_test)=keras.datasets.fashion_mnist.load_data()
```

```python
# shape of Train dataset
X_train.shape,y_train.shape
```

```python
# shape of Test dataset
X_test.shape,y_test.shape
```

```python
X_train[0]
```

```python
y_train[0]
```

```python
# Define the class label
class_labels=["T-shirt/Top","Trouser",'Pullover',"Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankel boot"]

'''
0=>T-shirt/Top
1=>Trouser
3=>Pullover
4=>Dress
5=>Coat
6=>Sandal
7=>Shirt
8=>Sneaker
9=>Ankel boot '''
```

```python
# Show The Firdt object
plt.imshow(X_train[0],cmap="Greys")
plt.show()
```

```python
# show the first 25 Images in dataset
plt.figure(figsize=(16,16))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(X_train[i],cmap="Greys")
    plt.axis('off')
    plt.title(class_labels[y_train[i]]+"="+str(y_train[i]),fontsize=20)   
    
    
```

### Feature Scaling

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train[0]
```

## Build Neural Network

```python
model=keras.models.Sequential([
                         keras.layers.Flatten(input_shape=[28,28]),
                         keras.layers.Dense(units=32,activation='relu'),
                         keras.layers.Dense(units=10,activation='softmax')
])
```

```python
# Summery of model
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=1)
```

```python
model.fit(X_train,y_train,epochs=10)
```

#### Training Model Accuracy is 89%

## Test and Evaluate Neural Network Model

```python
model.evaluate(X_test,y_test)
```

#### Training Model Accuracy is 89%

```python
y_pred=model.predict(X_test)
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,[np.argmax(i) for i in y_pred])
```

```python
# Heatmap
plt.figure(figsize=(16,16))
sns.heatmap(cm,annot=True,fmt="d")
plt.show()
```

## Classfication Report

```python
from sklearn.metrics import classification_report
```

```python
cr=classification_report(y_test,[np.argmax(i) for i in y_pred],target_names=class_labels,)
```
